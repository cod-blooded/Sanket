```python
import os
import cv2

DATA_PATH = './motion_data'

actions = ["HELLO", "YES", "NO", "THANKS", "OK"]

no_sequences = 60
sequence_length = 30

cap = cv2.VideoCapture(0)

exit_flag = False

for action in actions:
    if exit_flag:
        break

    print(f"\n➡️ Get ready for: {action}")

    # 🔹 WAIT FOR START (ONCE PER WORD)
    while True:
        ret, frame = cap.read()
        frame = cv2.flip(frame, 1)

        cv2.putText(frame, f'Action: {action}',
                    (20, 40), cv2.FONT_HERSHEY_SIMPLEX,
                    0.8, (0, 255, 0), 2)

        cv2.putText(frame, 'Press S to START | ESC to QUIT',
                    (20, 80), cv2.FONT_HERSHEY_SIMPLEX,
                    0.7, (255, 255, 0), 2)

        cv2.imshow('Collecting Motion Data', frame)

        key = cv2.waitKey(1) & 0xFF

        if key == ord('s'):
            break
        elif key == 27:
            print("🛑 Exiting...")
            exit_flag = True
            break

    if exit_flag:
        break

    # 🔹 SINGLE COUNTDOWN (ONLY ONCE)
    for i in range(3, 0, -1):
        ret, frame = cap.read()
        frame = cv2.flip(frame, 1)

        cv2.putText(frame, f'{action}',
                    (20, 40), cv2.FONT_HERSHEY_SIMPLEX,
                    0.8, (0, 255, 0), 2)

        cv2.putText(frame, f'Starting in {i}',
                    (200, 200), cv2.FONT_HERSHEY_SIMPLEX,
                    1.5, (0, 0, 255), 3)

        cv2.imshow('Collecting Motion Data', frame)

        key = cv2.waitKey(1000) & 0xFF
        if key == 27:
            print("🛑 Exiting...")
            exit_flag = True
            break

    if exit_flag:
        break

    # 🔹 CONTINUOUS CAPTURE (NO TIMER BETWEEN SEQUENCES)
    for sequence in range(no_sequences):
        if exit_flag:
            break

        sequence_path = os.path.join(DATA_PATH, action, str(sequence))
        os.makedirs(sequence_path, exist_ok=True)

        print(f"Recording {action} | Sequence {sequence}")

        for frame_num in range(sequence_length):
            ret, frame = cap.read()
            frame = cv2.flip(frame, 1)

            cv2.putText(frame,
                        f'{action} | Seq {sequence} | Frame {frame_num}',
                        (20, 40), cv2.FONT_HERSHEY_SIMPLEX,
                        0.7, (0, 255, 0), 2)

            cv2.imshow('Collecting Motion Data', frame)

            cv2.imwrite(
                os.path.join(sequence_path, f"{frame_num}.jpg"),
                frame
            )

            key = cv2.waitKey(30) & 0xFF
            if key == 27:
                print("🛑 Exiting...")
                exit_flag = True
                break

cap.release()
cv2.destroyAllWindows()
```
